```python
##Splitting into continuous and categorical features    
conFeat = ['age', 'fnlwgt', 'capital-gain', 'capital-loss', 'hours-per-week']
catFeat = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'target']
```

```python
##creating headers for my output csv files
outHead1 = ['Count', 'Miss %', 'Card.', 'Min', '1st Qrt.', 'Mean', 'Median', '3rd Qrt', 'Max', 'Std. Dev.']
outHead2 = ['Count', 'Miss %', 'Card.', 'Mode', 'Mode Freq', 'Mode %', '2nd Mode', '2nd Mode Freq', '2nd Mode %']
```

```python
##getting ready to export my continous features
conOut=pd.DataFrame(index=conFeat, columns=outHead1)
columns = dataset[conFeat]
conOut.index.name = 'FEATURENAME'
```

```python
## Setting up correct header names

#count

count = columns.count()
conOut[outHead1[0]]=count

#missing percentage is always 0

conOut[outHead1[1]]= 0.00

#cardinality
conOut[outHead1[2]]= columns.nunique()

conOut[outHead1[3]]= columns.min()

#first quartile
conOut[outHead1[4]]= columns.quantile(0.25)

conOut[outHead1[5]]= columns.mean()

conOut[outHead1[6]]= columns.median()

#third quartile
conOut[outHead1[7]]= columns.quantile(0.75)

conOut[outHead1[8]]= columns.max()

conOut[outHead1[9]]= columns.std()


```

```python
##getting ready to export categorical 
catOut=pd.DataFrame(index=catFeat, columns=outHead2)
columns = dataset[catFeat]
catOut.index.name = 'FEATURENAME'
```

```python
## finding mode, mode freq , mode % 2nd mode 2nd mode freq and 2nd mode %
dataset['workclass'].value_counts()

##creating arrays
mode_array = ['']*len(catFeat)
mode2_array=['']*len(catFeat)

mode_freq_array=['']*len(catFeat)
mode2_freq_array=['']*len(catFeat)

mode_freq_array_prcnt=['']*len(catFeat)
mode2_freq_array_prcnt=['']*len(catFeat)


for col in columns:
        values = columns[col].value_counts()
        index = catFeat.index(col)
        
        ##mode is the first value of the list of value counts
        mode=values.index[0]
        
        ##2nd mode is 2nd value
        mode2=values.index[1]
        
        ##entering modes into array
        mode_array[index]=mode
        mode2_array[index]=mode2
        
        ##mode frequencies
        modeCount = values.loc[mode]
        mode2Count =values.loc[mode2]
        
        mode_freq_array[index] = modeCount
        mode2_freq_array[index] = mode2Count
        
        ##mode frequency percentage = mode frequency/count*100
        ##count is always 30940
        
        ##Since all mode counts are above 0 no need for error checking
        mode_freq_array_prcnt[index]= (modeCount/30940)*100
        mode2_freq_array_prcnt[index]= (mode2Count/30940)*100
        
```

```python
##find missing percentage
prcnt_miss=['']*len(catFeat)

missing = columns.isna()
prcnt_num_missing=missing.sum()
prcnt_num_missing=prcnt_num_missing/ len(dataset)
prcnt_num_missing*100

```

```python
##enter missing percent
prcnt_miss[0]=5.6
prcnt_miss[1]=0
prcnt_miss[2]=0
prcnt_miss[3]=5.6
prcnt_miss[4]=0
prcnt_miss[5]=0
prcnt_miss[6]=0
prcnt_miss[7]=1.8
prcnt_miss[8]=0
```

```python

##count is always 30940
catOut[outHead2[0]] = 30940

catOut[outHead2[1]] = prcnt_miss

catOut[outHead2[2]] = columns.nunique()

catOut[outHead2[3]] = mode_array

catOut[outHead2[4]] = mode_freq_array

catOut[outHead2[5]] = mode_freq_array_prcnt

catOut[outHead2[6]] = mode2_array

catOut[outHead2[7]] = mode2_freq_array

catOut[outHead2[8]] = mode2_freq_array_prcnt

```

```python
###Exporting to csv

conOut.to_csv("./data/C17478104CONT.csv")
catOut.to_csv("./data/C17478104CAT.csv")
```
